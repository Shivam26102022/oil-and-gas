```python
## just put the non-replicated columns into the db (decided schema in db-create)
scout = scout[['NDIC File No:', 'API No:', 'Status Date:', 'Wellbore type:', 
                 'Latitude:', 'Longitude:', 'Total Depth:']]
```

```python
scout.columns
```

```python
import numpy as np
# form the Lat Lon into postgis points

def form_pt(data):
    pt = 'Point(%s %s)' % (data['Longitude:'], data['Latitude:'])
    return pt
```

```python
scout['pt'] = scout.apply(form_pt, axis=1)
scout['pt'] = scout.pt.apply(lambda x: '' if x == 'Point( )' else x)
scout[1000:1010]
scout.head()
```

```python
scout[1000:1010]
```

```python
import uuid
## make uuid for csv import into db

uids = [[x, str(uuid.uuid4())] for x in list(scout['NDIC File No:'])]
uidDF = pd.DataFrame(uids, columns = ['NDIC File No:', 'uid'])
```

```python
scout = pd.merge(scout, uidDF, right_on = 'NDIC File No:', 
                     left_on = 'NDIC File No:')
scout.head()
```

```python
del scout['Latitude:']
del scout['Longitude:']
scout.columns = ['ndic', 'api', 'status_date', 'wellbore_type', 'total_depth', 'pt', 'uid']
```

```python
scout.head()
```

```python
poolDF = prod[['Pool oil and/or gas was produced from', 'Element']].drop_duplicates()
poolDF.info()
```

```python

scout.info()
scoutDD = scout[scout.ndic.notnull()]
scoutDD['ndic'] = scoutDD.ndic.astype(int)
poolDF.columns = ['pool', 'ndic']
scoutTemp = pd.merge(scoutDD, poolDF, left_on = 'ndic', right_on = 'ndic', how = 'left')
scoutTemp.info()
```

```python
print scoutTemp.ndic.max(), scoutTemp.ndic.min()
```

```python
## look for duplicate data in the dataset
ndicDuplicate = scoutTemp.groupby('ndic').count()['uid'] > 1 
print scoutTemp[scoutTemp.ndic.isin(ndicDuplicate[ndicDuplicate].index)]
```
